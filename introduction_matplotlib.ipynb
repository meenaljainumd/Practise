```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
fig, ax = plt.subplots()
x= np.linspace(-1,1)
ax.plot(x, np.log(x));
```

```python
fig, ax = plt.subplots(nrows =2, ncols =2, figsize = (10,5))
ax[0,0].plot(x, np.exp(-x))
y2=  np.exp(-x)+ 1
ax[0,1].plot(x,y2)
y3 = 1/y2
ax[1,0].plot(x,y3)
```
